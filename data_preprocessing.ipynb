Modularized code:

```python
def apply_lowercase(text):
    return text.lower()
```

```python
def make_tokens(text):
    return word_tokenize(text)
```

```python
def remove_special_chars(tokens):
    return [re.sub(r'[^a-zA-Z]', '', token) for token in tokens]
```

```python
def apply_spell_check(tokens):
    spell = SpellChecker()
    return [spell.correction(token) for token in tokens if spell.correction(token) is not None]
```

```python
def remove_stop_words(tokens):
    stop_words = set(stopwords.words('english'))
    return [token for token in tokens if token not in stop_words]
```

```python
def apply_stemmer(tokens):
    stemmer = PorterStemmer()
    return [stemmer.stem(token) for token in tokens]
```

```python
def apply_lemmatizer(tokens):
    lemmatizer = WordNetLemmatizer()
    return [lemmatizer.lemmatize(token) for token in tokens]
```

```python
def remove_whitespace(text):
    return ' '.join(text.split())
```

```python
def handle_contractions(text):
    return contractions.fix(text)
```

```python
def apply_tokenization(text):
    tokens = make_tokens(text)
    tokens = remove_special_chars(tokens)
    tokens = apply_spell_check(tokens)
    tokens = remove_stop_words(tokens)
    # tokens = apply_stemmer(tokens)
    tokens = apply_lemmatizer(tokens)
    return tokens
```

```python
def preprocess_text_modularize(text):
    '''
    Preprocess text by doing the following: Lowercase text, Tokenization, Remove Special Characters,
    Remove Stop Words, Stemming & Lemmatization, Spell Check, Remove Whitespace, & Handle contractions
    '''
    text = apply_lowercase(text)
    text = handle_contractions(text)
    text = remove_whitespace(text)
    tokens = apply_tokenization(text)
    text = ' '.join(tokens)
    return text
```

```python
# all preprocessing steps except stemming

preprocessed_path2 = '/content/drive/MyDrive/CS505 Datasets/project_data/preprocessed_data_no_stemming'
os.makedirs(preprocessed_path2, exist_ok=True)
```

```python
for filename in os.listdir(dataset_path):
    if filename.endswith(".json"):
        file_path = os.path.join(dataset_path, filename)

        with open(file_path, 'r') as file:
            try:

                file_content = file.read()
                print(f"File Content of {filename}: {file_content}")

                data = json.loads(file_content)
                text_to_preprocess = data.get("Text", "")

                preprocessed_text = preprocess_text_modularize(text_to_preprocess)

                new_filename = os.path.splitext(filename)[0] + "_preprocessed3.json"
                new_file_path = os.path.join(preprocessed_path2, new_filename)

                data["Text"] = preprocessed_text

                with open(new_file_path, 'w') as new_file:
                    json.dump(data, new_file, indent=4)

            except json.JSONDecodeError as e:
                print(f"Error decoding JSON in {filename}: {e}")
```
